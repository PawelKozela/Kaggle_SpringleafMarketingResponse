```python
def calculate_correls_for_var(var, default_values):
    
    X_tmp = pd.DataFrame(X_train_raw[var])
    X_tmp['toto'] = 1
    X_tmp['target'] = Y_train['target']
    X_tmp.index = Y_train.index
    pivot_tmp = pd.pivot_table(X_tmp, index=var, columns='target', values='toto', aggfunc='count')
    pivot_tmp['Ratio'] = np.round(100 * pivot_tmp.iloc[:, 1] / (pivot_tmp.iloc[:, 0] + pivot_tmp.iloc[:, 1]), 0)
    filters = (~pivot_tmp.Ratio.isnull())
    default_value_filters = (~pivot_tmp.index.isin(default_values))

    ar_1 = pivot_tmp[filters].index
    ar_2 = pivot_tmp[filters]['Ratio']
    
    ar_1_d = pivot_tmp[filters & default_value_filters].index
    ar_2_d = pivot_tmp[filters & default_value_filters]['Ratio']
    
    print var, 100 * np.corrcoef(ar_1, ar_2)[0,1], 100 * np.corrcoef(ar_1_d, ar_2_d)[0,1]
    #pivot_tmp[filters & default_value_filters]['Ratio'].plot(figsize=(30,10))
```
